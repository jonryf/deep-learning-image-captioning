```python
from Decoder import Decoder
from Encoder import Encoder
from evaluate_captions import evaluate_captions
from runner import Runner
from settings import LSTM_HIDDEN_SIZE, EMBEDDED_SIZE, CAPTIONS_DIR, IMAGES_DIR, NUM_WORKERS, SHUFFLE_DATA, BATCH_SIZE, VALIDATION_SIZE, TEMPERATURE
from utils import load_datasets, select_ann_ids, load_image_ids, get_device

from data_loader import get_loader
```

```python
training_ids = select_ann_ids(load_image_ids(True), "captions_train2014.json")

validation_ids = training_ids[:int(len(training_ids) * (VALIDATION_SIZE))]
training_ids = training_ids[int(len(training_ids) * (VALIDATION_SIZE)):]
testing_ids = select_ann_ids(load_image_ids(False), "captions_val2014.json")

# load data and transform images
train_dataset = get_loader(IMAGES_DIR + "/train/", CAPTIONS_DIR + "/captions_train2014.json", training_ids,
                           BATCH_SIZE, SHUFFLE_DATA,
                           NUM_WORKERS)
```

```python
vocabulary_size = len(train_dataset.dataset.vocab.word2idx)

computing_device = get_device()

encoder = Encoder(EMBEDDED_SIZE).to(computing_device)
decoder = Decoder(EMBEDDED_SIZE, LSTM_HIDDEN_SIZE, vocabulary_size).to(computing_device)

runner = Runner(encoder, decoder, train_dataset, train_dataset, train_dataset)
runner.train()
```
